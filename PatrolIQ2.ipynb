Clustering

```python
features=df[numerical_col]
```

```python
features.head()
```

```python
features.drop(columns=['ID','Beat','District','Ward','Community Area'],inplace=True)
```

```python
features.head()
```

```python
#Normalization
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X=scaler.fit_transform(features)
X[:5]
```

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans,DBSCAN,AgglomerativeClustering
from sklearn.metrics import silhouette_score,davies_bouldin_score
```

```python
inertia_list=[]
k_range=range(2,30)

for k in k_range:
    kmeans=KMeans(n_clusters=k,random_state=42)
    kmeans.fit(X)
    inertia_list.append(kmeans.inertia_)

#plot elbow curve
plt.figure(figsize=(6,4))
plt.plot(k_range,inertia_list,'bo-')
plt.xlabel('No. of clusters')
plt.ylabel('Inertia')
plt.title('K-means Elbow Method')
plt.show()    
```

```python

kmeans = KMeans(n_clusters=4, random_state=42)
kmeans_labels = kmeans.fit_predict(X)

df['KMeans_Cluster'] = kmeans_labels
df['KMeans_Cluster'].head()
```

```python
k_means_sil=silhouette_score(X,kmeans_labels)
k_means_davis=davies_bouldin_score(X,kmeans_labels)
print("KMeans - Silhouette:", k_means_sil, "DB Index:", k_means_davis)
```

```python
plt.figure(figsize=(8,6))
scatter = plt.scatter(df['Longitude'], df['Latitude'], c=df['KMeans_Cluster'], s=10, cmap='tab10')

plt.title("KMeans Clusters â€“ Crime Hotspots", fontsize=16)
plt.xlabel("Longitude", fontsize=12)
plt.ylabel("Latitude", fontsize=12)

cbar = plt.colorbar(scatter)
cbar.set_label("Cluster ID")

plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import DBSCAN

db = DBSCAN(eps=0.19, min_samples=10).fit(X)
db_labels = db.labels_

df['DBSCAN_Cluster'] = db_labels
df.head()
```

```python
# Evaluation (ignore noise points - label = -1)
mask = db_labels != -1
dbscan_sil = silhouette_score(X[mask], db_labels[mask])
dbscan_db = davies_bouldin_score(X[mask], db_labels[mask])
print("DBSCAN - Silhouette:", dbscan_sil, "DB Index:", dbscan_db)
```

```python
plt.figure(figsize=(8,6))
scatter = plt.scatter(df['Longitude'], df['Latitude'], c=df['DBSCAN_Cluster'], s=10, cmap='viridis')

plt.title("DBSCAN â€“ Geographic Crime Density Clusters", fontsize=16)
plt.xlabel("Longitude", fontsize=12)
plt.ylabel("Latitude", fontsize=12)

cbar = plt.colorbar(scatter)
cbar.set_label("Cluster ID ( -1 = Noise )")

plt.tight_layout()
plt.show()
```

```python
df['DBSCAN_Cluster'].value_counts()
```

```python
from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters=3,linkage='ward')
hc_labels = hc.fit_predict(X)

df['HC_Cluster'] = hc_labels
df.head()
```

```python
# Evaluation
hier_sil = silhouette_score(X, hc_labels)
hier_db = davies_bouldin_score(X, hc_labels)
print("Hierarchical - Silhouette:", hier_sil, "DB Index:", hier_db)
```

```python
plt.figure(figsize=(8,6))
scatter = plt.scatter(df['Longitude'], df['Latitude'], c=df['HC_Cluster'], s=10, cmap='Set1')

plt.title("Hierarchical Clustering â€“ Crime Zones", fontsize=16)
plt.xlabel("Longitude", fontsize=12)
plt.ylabel("Latitude", fontsize=12)

cbar = plt.colorbar(scatter)
cbar.set_label("Cluster ID")

plt.tight_layout()
plt.show()
```

```python
# ---------------------------
# 9. COMPARE AND CHOOSE BEST ALGORITHM
# ---------------------------
scores = {
    'KMeans': k_means_sil,
    'DBSCAN': dbscan_sil,
    'Hierarchical': hier_sil
}

best_algo = max(scores, key=scores.get)
print("Best Clustering Algorithm:", best_algo)
```

```python
#Dimensionality Reduction-PCA
X_all=df1[num_col].copy()
scaler=StandardScaler()
X_all_scaled=scaler.fit_transform(X_all)
pca=PCA(n_components=3)
X_pca=pca.fit_transform(X_all_scaled)
print("Explained Variance Ratios:", pca.explained_variance_ratio_)
print("Cumulative Variance:", np.cumsum(pca.explained_variance_ratio_))
```

```python
# ---------------------------
# 7. TOP FEATURES FROM PCA
# ---------------------------
# Absolute value of components to get feature importance
pca_components = pd.DataFrame(abs(pca.components_), columns=num_col)
feature_importance = pca_components.sum().sort_values(ascending=False)
print("Top Features Driving Crime Patterns:\n", feature_importance)
```

```python
plt.figure(figsize=(7,5))
plt.bar(range(1,4), pca.explained_variance_ratio_, tick_label=['PC1','PC2','PC3'])

plt.title("PCA â€“ Explained Variance per Component", fontsize=16)
plt.xlabel("Principal Components", fontsize=12)
plt.ylabel("Variance Ratio", fontsize=12)

plt.tight_layout()
plt.show()
```
